Python's `numpy.meshgrid` is ideal for doing quantitative work in images with Cartesian and polar coordinates. Images can be plotted with `matplotlib.image`, but the coordinates will be off by one (or worse!) unless you exercise care with the `extent` tuple!

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg


# Define pixel coordinates for producing test images
Nx = 7
Ny = 4
x = range(-Nx,Nx)
y = range(-Ny,Ny)
xs,ys = np.meshgrid(x,y)
rs = np.sqrt(xs**2 + ys**2) # radial coordinate
# Specify correct extent for plt.imshow(... origin='lower')
extent = (x[0]-0.5,x[-1]+0.5,y[0]-0.5,y[-1]+0.5)

# Demonstrate that the extent is specified correctly
imgplot = plt.imshow(rs, cmap='gray', extent=extent, 
                     aspect='equal', origin='lower')
plt.title("My preferred option is origin='lower'")
plt.show()

# Demonstrate that if I use the default ('upper') origin with imshow,
# the last 2 entries in the extent need to be flipped:
imgplot = plt.imshow(rs, cmap='gray', extent=extent, 
                     aspect='equal')
plt.title("Incorrect!")
plt.show()

extent2 = (x[0]-0.5,x[-1]+0.5,y[-1]+0.5,y[0]-0.5)
imgplot = plt.imshow(rs, cmap='gray', extent=extent2, 
                     aspect='equal')
plt.title("origin='upper' (default)")
plt.show()


```
